For the dict *exp_dcit*, the keys are the name of experiments and the values are the correspoding dictionaries with:
* *dir*: the directory of experment results
* *seeds*: the number of repeated experiments, i.e., the maximum of exp_index
* *color*: the correspoding color in the plot
* *label*: the corresponding label in the plot

```python
env_dir = "../attention_allocation_experiment/experiments/harder_env/"

exp_dict = {
    "G-PPO":{"dir":env_dir+"GPPO/lr_1e-05", 
             "seeds":3, 
             "color":"green", 
             "label":"G-PPO"},
    "R-PPO":{"dir":env_dir+"RPPO/lr_1e-05", 
             "seeds":3, 
             "color":"blue", 
             "label":"R-PPO"},
    "A-PPO":{"dir":env_dir+"APPO/lr_1e-05", 
             "seeds":3,
             "color":"purple", 
             "label":"A-PPO"},
    "ours":{"dir":env_dir+"ELBERT/smooth_20.0/alpha_20000.0_lr_1e-05", 
            "seeds":3, 
            "color":"red", 
            "label":"ours"}
}
```

```python
fig_reward_plot = load_and_plot(exp_dict, plot_type='Rewards', ylim=[-5.5, -2], xlim=[0, 5e6], timesteps=245, legend_loc='lower right')
```
